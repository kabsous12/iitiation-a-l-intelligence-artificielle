```python
import numpy as np 
import pandas as pd 
```

```python
All_data = pd.read_csv("Data/USA_Housing.csv")
```

```python
All_data
```

```python
data = pd.DataFrame({'x':All_data.iloc[:,0],'y':All_data.iloc[:,5]})
data
```

```python
#analyse des données 
data.isnull().sum()
```

```python
data.info()
```

```python
data.head(4)
```

```python
x = data.iloc[:,:-1].values
y= data.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
#X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size = 0.25, random_state= 40)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', x_test.shape)
```

```python
# importer le modèle de la linear regression  
from sklearn.linear_model import LinearRegression 
```

```python
modele_exercice_test=LinearRegression()
```

```python
# entrainement du modèle 
modele_exercice_test.fit(x_train, y_train)
```

```python
y_scores_predict = modele_exercice_test.predict(x_test)
```

```python
y_scores_predict
```

```python
resultats = pd.DataFrame({'Area income':y_test,'Price':y_scores_predict})
```

```python
resultats
```

```python
# importer matplotlib 
import matplotlib.pyplot as plt 

plt.scatter(x_train, y_train, color="blue",label= 'train data')
plt.scatter(x_test, y_test, color="yellow",label= 'test data')
plt.plot(x_test, y_scores_predict, color='red', label ='ligne de la regression')
plt.legend()
plt.show()
```

```python
resultats = pd.DataFrame({'y_score_real':y_test,'y_scores_predict':y_scores_predict,'soustraction': y_test-y_scores_predict})
resultats
```
